### El gobierno tiene una pagina con datasets de distintas areas. Estoy importando directamente desde ahi para mantener actualizado el analisis.


```python
import pandas as pd

url_obras_soc = 'http://datos.salud.gob.ar/dataset/0fe5fc87-d980-41c2-8e71-3a7d0964350c/resource/7486a94a-047f-4e19-a742-5b6f2db9c854/download/padron-de-beneficiarios-de-obras-sociales-nacionales-20181217.csv'
url_prepagas = "http://datos.salud.gob.ar/dataset/017cd909-3534-4e11-9b41-cdce8f13276a/resource/7cd31f90-6ce2-4ed9-b1c4-bf454479090e/download/cantidad-de-usuarios-de-empresas-de-medicinas-prepagas-20181220.csv"

obras_soc = pd.read_csv(url_obras_soc)
prepagas = pd.read_csv(url_prepagas)
```

### Creo que la segunda opcion es la mas probable. Creo que no tengo forma de saber esto. Voy a calcular que porcentaje de los usuarios de medicina prepaga en relacion a los de obras sociales para ver si es relevante.




```python
df["procentaje_obras"] = df["obras_sociales"]/df["cant_habitantes"]*100

df["procentaje_prepagas"] = df["prepagas"]/df["cant_habitantes"]*100

df["rel_prepagas_obras_soc"] = df["prepagas"]/df["obras_sociales"]*100

df
```
